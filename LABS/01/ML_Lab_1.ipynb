## TASK 3:

A. Perform Graphical Visualization

B. Display a correlation graph

C. Display the distribution of Age and Frequent Cold

D. Find out outliers in Obesity feature by using Graphical Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

#B
plt.figure(figsize=(15,10))
sns.heatmap(dataset.corr(numeric_only=True), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap of Features")
plt.show()
```

```python
#C
plt.figure(figsize=(10,6))
sns.scatterplot(x=dataset['Age'], y=dataset['Frequent Cold'])
plt.title("Distribution of Age vs Frequent Cold")
plt.xlabel("Age")
plt.ylabel("Frequent Cold")
plt.show()
```

```python
#D
plt.figure(figsize=(8,6))
sns.boxplot(x=dataset['Obesity'])
plt.title("Outlier Detection in Obesity Feature")
plt.show()
```
